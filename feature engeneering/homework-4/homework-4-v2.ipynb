```python
import pandas as pd
import datetime
import pandas_profiling
```

```python
data = pd.read_csv('data/aac_shelter_outcomes.csv')
data.head()
```

```python
data.info()
```

```python
data.drop(['animal_id'], axis=1, inplace=True)
data.fillna('Other', inplace=True)
data.head()
```

```python
def monthbetween(row):
    format = '%Y-%m-%dT%H:%M:%S'
    d1 = datetime.datetime.strptime(row['date_of_birth'], format)
    d2 = datetime.datetime.strptime(row['datetime'], format)
    return (d2.year - d1.year) * 12 + d2.month - d1.month+1
```

```python
data['age'] = data.apply(monthbetween, axis=1)
data['monthout'] = data['datetime'].apply(lambda x: int(x[5:7]))
data.head()
```

```python
data[data['datetime'] != data['monthyear']]
```

```python
data.drop(['age_upon_outcome', 'datetime', 'monthyear', 'date_of_birth'], axis=1, inplace=True)
data.head()
```

```python
data.info()
```

```python
def targetenc(x):
    if x == 'Transfer':
        return 1
    elif x == 'Adoption':
        return 2
    else:
        return 3
```

```python
data['outcome_type_enc'] = data['outcome_type'].apply(targetenc)
data.head()
```

```python
def colorenc(x):
    if x == 'Black/White':
        return 0
    elif x == 'Black':
        return 1
    elif x == 'White':
        return 2
    elif 'Black' in x:
        return 3
    elif 'White' in x:
        return 4
    else:
        return 5
```

```python
data['color_enc'] = data['color'].apply(colorenc)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['animal_type_enc'] = le.fit_transform(data.animal_type)
data.head()
```

```python
def breedenc(x):
    if 'Mix' in x:
        return 0
    else:
        return 1
```

```python
data['breed_enc'] = data['breed'].apply(breedenc)
data.head()
```

```python
data.sex_upon_outcome.unique()
```

```python
def sexenc(x):
    if 'Male' in x:
        return 1
    if 'Female' in x:
        return 0
    else:
        return 3
    
def intenc(x):
    if 'Intact' in x:
        return 1
    if 'Spayed' or 'Neutered' in x:
        return 0
    else:
        return 3
```

```python
data['sex_enc'] = data['sex_upon_outcome'].apply(sexenc)
data['int_enc'] = data['sex_upon_outcome'].apply(intenc)
data.head(10)
```

```python
data.drop(['animal_type', 'breed', 'color', 'name', 
                                       'outcome_subtype', 'sex_upon_outcome', 'outcome_type'], axis=1, inplace = True)
```

```python
data.head()
```

```python
x = data.drop('outcome_type_enc', axis = 1)
y = data.outcome_type_enc
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
model_pred = model.predict(x_test)
print(classification_report(y_test, model_pred))
```
