```python
import pandas as pd
import datetime
import pandas_profiling
```

```python
data = pd.read_csv('data/aac_shelter_outcomes.csv')
data.head()
```

```python
data.info()
```

```python
data.drop(['animal_id'], axis=1, inplace=True)
data.fillna('Other', inplace=True)
data.head()
```

```python
def monthbetween(row):
    format = '%Y-%m-%dT%H:%M:%S'
    d1 = datetime.datetime.strptime(row['date_of_birth'], format)
    d2 = datetime.datetime.strptime(row['datetime'], format)
    return (d2.year - d1.year) * 12 + d2.month - d1.month+1
```

```python
data['age'] = data.apply(monthbetween, axis=1)
data['monthout'] = data['datetime'].apply(lambda x: int(x[5:7]))
data.head()
```

```python
data[data['datetime'] != data['monthyear']]
```

```python
data.drop(['age_upon_outcome', 'datetime', 'monthyear', 'date_of_birth'], axis=1, inplace=True)
data.head()
```

```python
data.info()
```

```python
def outcome(type):
    if type not in ['Adoption', 'Transfer']:
        return 'Other'
    else:
        return type
```

```python
data['outcome_type'] = data['outcome_type'].apply(outcome)
data.head(20)
```

```python
r = data.color.value_counts()[data.color.value_counts().values > 500].keys()
r
```

```python
data.sex_upon_outcome.value_counts()
```

```python
breeds = data.breed.value_counts()[data.breed.value_counts().values < 1000].keys()
colors = data.color.value_counts()[data.color.value_counts().values < 2000].keys()
names = data.name.value_counts()[data.name.value_counts().values < 100].keys()
```

```python
data.loc[data['breed'].isin(breeds), 'breed'] = 'Other'
data.loc[data['color'].isin(colors), 'color'] = 'Other'
data.loc[data['name'].isin(names), 'name'] = 'Other'
data.loc[data['sex_upon_outcome'] == 'Unknown', 'sex_upon_outcome'] = 'Other'
data.head(100)
```

```python
from sklearn.preprocessing import OneHotEncoder
one_enc = OneHotEncoder()
features = one_enc.fit_transform(data[['animal_type', 'breed', 'color', 'name', 
                                       'outcome_subtype', 'sex_upon_outcome']]).toarray()
fts = pd.DataFrame(features)
```

```python
data = pd.concat([data, fts], axis=1)
```

```python
targets = one_enc.fit_transform(data[['outcome_type']]).toarray()
targets_columns = one_enc.categories_[0]
tgs = pd.DataFrame(targets, columns=targets_columns)
tgs
```

```python
data = pd.concat([data, tgs], axis=1)
data.head()
```

```python
data.drop(['animal_type', 'breed', 'color', 'name', 
                                       'outcome_subtype', 'sex_upon_outcome', 'outcome_type'], axis=1, inplace = True)
```

```python
data.head()
```

```python
x = data.iloc[:,:82]
y = data.iloc[:,82:]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
model_pred = model.predict(x_test)
print(classification_report(y_test, model_pred))
```
