### Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
norm_rv1 = sts.norm(loc=1, scale=5)
norm_rv2 = sts.norm(loc=3, scale=8)
norm_rv3 = sts.norm(loc=5, scale=1)
norm_rv4 = sts.norm(loc=3, scale=3)
norm_rv5 = sts.norm(loc=2, scale=-5)

x = np.linspace(-20,20,100)

pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
plt.plot(x, pdf5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
expon_rv1 = sts.expon(0,1)
expon_rv2 = sts.expon(0.5)
expon_rv3 = sts.expon(2)
expon_rv4 = sts.expon(1)
expon_rv5 = sts.expon(3)

x = np.linspace(0,5,100)

cdf1 = expon_rv1.cdf(x)
cdf2 = expon_rv2.cdf(x)
cdf3 = expon_rv3.cdf(x)
cdf4 = expon_rv4.cdf(x)
cdf5 = expon_rv5.cdf(x)

plt.plot(x, cdf1)
plt.plot(x, cdf2)
plt.plot(x, cdf3)
plt.plot(x, cdf4)
plt.plot(x, cdf5)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
t_rv1 = sts.t(0.2)
t_rv2 = sts.t(0.5)
t_rv3 = sts.t(0.9)
t_rv4 = sts.t(5)
t_rv5 = sts.t(10)

x = np.linspace(-5,5,100)

cdf1 = t_rv1.cdf(x)
cdf2 = t_rv2.cdf(x)
cdf3 = t_rv3.cdf(x)
cdf4 = t_rv4.cdf(x)
cdf5 = t_rv5.cdf(x)

plt.plot(x, cdf1)
plt.plot(x, cdf2)
plt.plot(x, cdf3)
plt.plot(x, cdf4)
plt.plot(x, cdf5)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

### Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
def entropy(r, y, g):

    time = r + y + g

    pr = r/time
    py = y/time
    pg = g/time

    entr = -math.log2(pr)
    enty = -math.log2(py)
    entg = -math.log2(pg)
    
    return (entr, enty, entg)
```

```python
entropy(54,4,28)
```

```python
entropy(28,28,28)
```
