## Уровень 0

посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
from sympy import diff, symbols, cos, sin
```

```python
# 1st derivative

x = symbols('x')
diff(x**5 + 4*sin(2*x) + cos(3*x + 3)).evalf(subs={'x':1})
```

```python
# 2nd derivative

diff(diff(x**5 + 4*sin(2*x) + cos(3*x + 3))).evalf(subs={'x':1})
```

## Уровень 1

Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
f = (sin(2*x + 1))**5
d1 = diff(f)
d2 = diff(d1)

from sympy.plotting import plot
plot(d1, (x, -5, 5))
plot(d2, (x, -5, 5))
```
