```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
dt = pd.read_csv('adult.csv')
dt.head()
```

```python
dt.info()
```

```python
dt.groupby('capital-loss').count()
```

```python
'''
Так как основная цель домашки - оработать использование линейных моделей, а не максимальная точность, фича-инжиниринг будет
простым: двоичные категориальные значения делаем label, остальные - onehot
'''

onehot = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'native-country']
label = dt[['income', 'gender']]
```

```python
le = LabelEncoder()

for column in label:
    dt[column] = le.fit_transform(dt[column])
dt.head()
```

```python
dt = pd.get_dummies(data=dt, columns=onehot)
```

```python
dt.head()
```

```python
# нормализуем числовые колонки
tonorm = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
from sklearn.preprocessing import MinMaxScaler 
mms = MinMaxScaler() 
for column in tonorm:
    dt[column] = mms.fit_transform(dt[[column]])

dt.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500)
```

```python
x = dt.drop('income', axis=1)
y = dt.income

model.fit(x, y)
```

```python
model.score(x, y)
```

```python
svc.fit(x,y)
```
