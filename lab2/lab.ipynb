```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in ['name', 'topleft', 'botright', 'mainhue']:
    data[i] = le.fit_transform(data[i])
    
data.head()
```

```python
# Сделать кросс-валидацию данных с использованием подхода K-fold (n_folds=3)
# Решить задачу мультиклассовой классификации и предсказать религию страны (religion) по деталям ее флага
# Проверить качество классификации с использованием следующих метрик: Accuracy, F1-Score, Precision, Recall

X = data.drop('religion', axis=1)
Y = data.religion

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
rf = RandomForestClassifier()
kn = KNeighborsClassifier()

from sklearn.model_selection import KFold
kf = KFold(n_splits=3)

from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
from statistics import mean
```

```python
import warnings
warnings.filterwarnings('ignore')

ac = []
f1 = []
re = []
pr = []
for train, test in kf.split(X):
    predictions = rf.fit(X.iloc[train], Y[train]).predict(X.iloc[test])
    ac.append(accuracy_score(Y[test], predictions))
    f1.append(f1_score(Y[test], predictions, average='macro'))
    re.append(recall_score(Y[test], predictions, average='macro'))
    pr.append(precision_score(Y[test], predictions, average='macro'))
    
print('accuracy_score = ', mean(ac))
print('precision_score = ', mean(pr))
print('recall_score = ', mean(re))
print('f1_score = ', mean(f1))
```
